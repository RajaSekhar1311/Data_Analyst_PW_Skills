**Conditional statements** in Python allow you to execute certain blocks of code based on specific conditions. These statements are fundamental for controlling the flow of your program.

Letâ€™s go through the main types of conditional statements with examples:

1. **if Statement**
The if statement executes a block of code if a specified condition is true.

**Example:**

```python
x = 10
if x > 5:
    print("x is greater than 5")
```

2. **if-else Statement**

The if-else statement executes one block of code if the condition is true, and another block if the condition is false.

**Example:**

```python
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

3. **if-elif-else Statement**

The if-elif-else statement allows you to check multiple conditions. The first condition that evaluates to true will have its corresponding block executed.

**Example:**

```python
x = 7
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but less than or equal to 10")
else:
    print("x is 5 or less")
```

4. **Nested if Statements**

You can nest if statements within other if statements to create more complex conditions.

**Example:**

```python
x = 8
if x > 5:
    if x < 10:
        print("x is between 5 and 10")
    else:
        print("x is 10 or more")
else:
    print("x is 5 or less")
```

**Key points:**

1. Conditions are evaluated to either **True** or **False**.

2. The code block within the **if** or **elif** statement is executed only if the corresponding condition is **True**.

3. The **else** block is executed if none of the previous conditions are **True**.

4. Multiple **elif** statements can be used to check for different conditions.

5. Indentation is crucial in Python to determine the scope of code blocks within conditional statements.

**Additional notes:**

* You can combine multiple conditions using logical operators (and, or, not).


* Python has built-in comparison operators (==, !=, <, >, <=, >=) for comparing values.
