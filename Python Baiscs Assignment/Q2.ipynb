Default keywords in Python are reserved words that have a special meaning and are an integral part of the syntax of the language. They cannot be used as labels (such as variable names or function names) because they are used to describe the structure and flow of a Python program.

Here are some basic functions and examples of default keywords in Python.

1. **Control Flow Keywords**

  **if, elif, else:** Used for conditional statements.

```python
x=10
if x > 0:
    print("Positive")
elif x == 0:
    print("Zero")
else:
    print("Negative")
```

2. **Looping Keywords:**

**for, while, break, continue:** Used for creating loops and controlling their execution.

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

3. **Function and Class Definition Keywords:**

**def, return, class:** Used to define functions and classes

```python
def greet(name):
    return ("Hello, {name}")

class Dog:
    def __init__(self, name):
        self.name = name
```

4. **Exception Handling Keywords:**

**try, except, finally, raise:** Used for handling exceptions.

```python
try:
    result = 10 / 00
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("Execution complete")
```

5. **Logical and Comparison Keywords:**

**and, or, not, is, in:** Used for logical operations and comparisons

```python
x=5
if x > 0 and x < 10:
    print("x is a single-digit positive number")
```
